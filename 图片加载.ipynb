```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import os
import cv2
import numpy as np
```

```python
# 打开一个jpg图像文件
im = Image.open(r"C:\Users\All\Pictures\test1.jpg")
# 获得图像尺寸:
w, h = im.size
print('Original image size: %sx%s' % (w, h))
plt.imshow(im)
plt.show()
# 缩放到25%:
im.thumbnail((w//4, h//4))
print('Resize image to: %sx%s' % (w//4, h//4))
# 把缩放后的图像用jpeg格式保存:
#im.save('thumbnail.jpg', 'jpeg')
plt.imshow(im)
plt.show()
```

```python
im = Image.open(r"C:\Users\All\Pictures\test.jpg")
# 应用模糊滤镜:
im2 = im.filter(ImageFilter.BLUR)
#im2.save('blur.jpg', 'jpeg')
plt.imshow(im2)
```
