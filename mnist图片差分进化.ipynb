```python
def Func(x):
    result=model.predict(x[:].reshape(1,784))
    return result[0][7]
```

```python
MAX_GENERATION = 2000  # 
GENERATION = 0  # 
L = 784  # 
N = 100  # 
SIGMA = 0.333  # 
CR = 0.1  # 
E = 0.15
T = 1
G = np.zeros(T)
for k in range(T):
        GENERATION = 0  # 
        BEST_X = 0
        FITS = np.zeros(N)
        FITS_U = np.zeros(N)
        V = np.zeros((N, L))
        U = np.zeros((N, L))
        # 1
        X = np.concatenate(x_des, axis=0)
        # 
        for i in range(N):
            FITS[i] = Func(X[i])
        #  
        BEST_X = X[np.argmin(FITS)]
        while GENERATION < MAX_GENERATION:
            if Func(BEST_X) < E:
                print(GENERATION)
                print(BEST_X)
                plt.imshow(BEST_X.reshape(28,28),cmap='gray')
                plt.show()
                break
            # 2
            for i in range(N):
                r = np.random.randint(1, N, 2)
                # V[i, :]
                # BEST_X
                # V[i, :] = BEST_X + SIGMA * (X[r[0], :] - X[r[1], :]) + SIGMA * (X[r[2], :] - X[r[3], :])
                # V[i, :] = BEST_X + SIGMA * ((X[r[0], :] - X[r[1], :]) + (X[r[2], :] - X[r[3], :]))
                # SIGMA 
                # SIGMA 
                V[i, :] = abs(BEST_X + SIGMA * (X[r[0], :] - X[r[1], :]))
                for j in range(784):
                    if V[i, j] > 1:
                        #print(V[i, j])
                        V[i, j] = 1-(V[i, j]-1)
                        #print(V[i, j])
            # print(V)
            # 3
            for i in range(N):
                jRand = np.floor(np.random.random() * L)
                for j in range(L):
                    # U[i,j]
                    # CR BEST_X
                    if np.random.random() > CR or j == jRand:
                        U[i, j] = X[i, j]
                    else:
                        U[i, j] = V[i, j]
                    #if np.random.random() < CR or j == jRand:
                        #U[i, j] = V[i, j]
                    #else:
                        #U[i, j] = X[i, j]
            # 4+
            for i in range(N):
                FITS[i] = Func(X[i])
                FITS_U[i] = Func(U[i])
                # 
                if FITS[i] >= FITS_U[i]:
                    X[i, :] = U[i, :]
                    FITS[i] = FITS_U[i]
                    # BEST_X
                    if FITS_U[i] < Func(BEST_X):
                        BEST_X = U[i, :]
            GENERATION += 1
            print("", Func(BEST_X))
        if GENERATION == MAX_GENERATION:
            print("")
        G[k] = GENERATION
       
print("", G)
print("", np.mean(G))
```
